Однако это число было бы некорректным.

Создайте в ваших данных о транзакциях признак `QuantityCanceled`. Этот признак будет указывать на количество отменённого впоследствии товара для каждой транзакции, на которую найдётся противоположная ей транзакция с возвратом. Для транзакций, для которых не было возвратов, признак будет равен 0. Для транзакций, для которых указано отрицательное количество товара, но в данных отсутствует контрагент, признак будет пустым.

В качестве вспомогательного инструмента мы подготовили для вас функцию `get_quantity_canceled()`. Эта функция принимает на вход таблицу с транзакциями и возвращает объект `Series` — столбец, в котором указано количество отменённого впоследствии товара для каждой транзакции. Если транзакция не имеет контрагента, этот признак помечается как `NaN`.

Отметим, что эта функция не учитывает сложный случай, когда количество отменённого товара в транзакции-возврате больше, чем количество товара, которое указано в любой из отдельных транзакций на покупку (это случай, когда клиент сделал несколько заказов одного и того же товара, а потом оформил возврат на все товары разом). При желании вы можете самостоятельно модифицировать функцию для предобработки, чтобы учесть этот пограничный случай.

*Осторожно! Поиск отменённых транзакций занимает много времени из-за большого количества строк в таблице. Выполнение следующего кода может занять до 20 минут в зависимости от мощности компьютера.*

```python
def get_quantity_canceled(data):
    """Функция для создания признака количества отменённых заказов. 
    Функция принимает на вход таблицу и возвращает столбец, в котором указано количество отменённого впоследствии товара 
    для каждой транзакции.
    Если транзакция с отрицательным количеством товара не имеет контрагента, данный признак помечается как NaN.

    Args:
        data (DataFrame): таблица с транзакциями

    Returns:
        Series: столбец с количеством отменённого товара
    """
    # Инициализируем нулями Series той же длины, что и столбцы таблицы.
    quantity_canceled = pd.Series(np.zeros(data.shape[0]), index=data.index)    
    negative_quantity = data[(data['Quantity'] < 0)].copy()
    for index, col in negative_quantity.iterrows():
        # Создаём DataFrame из всех контрагентов
        df_test = data[(data['CustomerID'] == col['CustomerID']) &
                       (data['StockCode']  == col['StockCode']) & 
                       (data['InvoiceDate'] < col['InvoiceDate']) & 
                       (data['Quantity'] > 0)].copy()
        # Транзация-возврат не имеет контрагента — ничего не делаем
        if (df_test.shape[0] == 0): 
            # Помечаем столбец как пропуск
            quantity_canceled.loc[index] = np.nan
        # Транзакция-возврат имеет ровно одного контрагента
        # Добавляем количество отменённого товара в столбец QuantityCanceled 
        elif (df_test.shape[0] == 1): 
            index_order = df_test.index[0]
            quantity_canceled.loc[index_order] = -col['Quantity']       
        # Транзакция-возврат имеет несколько контрагентов
        # Задаём количество отменённого товара в столбец QuantityCanceled для той транзакции на покупку,
        # в которой количество товара больше количества товаров в транзакции-возврате.
        elif (df_test.shape[0] > 1): 
            df_test.sort_index(axis=0 ,ascending=False, inplace = True)        
            for ind, val in df_test.iterrows():
                if val['Quantity'] < -col['Quantity']: 
                    continue
                quantity_canceled.loc[ind] = -col['Quantity']
                break    
    return quantity_canceled

data['QuantityCanceled'] = get_quantity_canceled(data)
```

Узнаем, сколько транзакций в данных не имеют контрагентов, и, если их количество невелико, удалим их из данных.

Когда мы разобрались с возвратами, удалим теперь те транзакции которые имеют отрицательное количество товара — они нам больше не нужны.

**Подсказка.** В качестве шаблона для поиска используйте строку '^[a-zA-Z]+'.

### 2.3. Построение RFM-таблицы и поиск RFM-выбросов

Мы добрались до самой интересной части нашей задачи. Нам предстоит сформировать признаки, на основе которых мы будем производить сегментацию клиентов.

Для этого давайте познакомимся с очень популярным методом для анализа потребительской ценности под названием RFM. 

<center> <img src=https://miro.medium.com/max/1400/1*uYQjy9SUjW7iWHc2gGanQQ.png align="right" width="400"/> </center>

Метод заключается в группировке клиентов на основе следующих параметров:
* Recency (Давность) — давность последней покупки клиента;
* Frequency (Частота) — общее количество покупок клиента;
* Monetary Value (Денежная ценность) — сколько денег потратил клиент.


Суть RFM-анализа состоит в том, что мы разделяем всех клиентов на группы в зависимости от того, как давно они сделали последнюю покупку, как часто покупали и насколько большой была сумма их заказов. 

Например, вот так может выглядеть интерпретация кластеров для случая RF-сегментации (анализа на основе давности и частоты заказов клиента):

<img src=https://retailrocket.ru/wp-content/uploads/2017/06/rfm-1.png>

Задача маркетологов — вести клиента в зону лояльных.

Мы можем рассчитать RFM-характеристики для каждого из клиентов в нашем датасете и на их основе с помощью методов кластеризации построить подобные сегменты клиентов, привязанные к нашим данным.

Чтобы получить RFM-таблицу, нам необходимо сгруппировать данные по идентификаторам клиента и рассчитать следующие  агрегированные характеристики:

* Recency для i-го клиента рассчитывается как разница между датой и временем последнего заказа и точкой отсчёта, переведённая в дни:
    $$t_0-max(t_{i1}, t_{i2},..., t_{iM})$$

    где $t_{ij}$ — дата и время совершения i-ым клиентом своей j-ой покупки.

    В качестве точки отсчёта $t_0$ берём дату на один день «старше», чем все наши данные. Это будет 10 декабря 2011 года (в формате datetime — '2011-12-10 00:00:00').

* Frequency рассчитывается как общее количество уникальных заказов, которые совершил i-ый клиент.
* Monetary Value рассчитывается как общая сумма денег, которую i-ый клиент потратил на наши товары (с учётом возвратов).

```python
# Для начала рассчитаем Recency, Frequency, Monetary value для каждого клиента
# Но перед этим укажем нашу точку отсчета

t0 = datetime(2011, 12, 10)

recency = data_cleaned.groupby('CustomerID')['InvoiceDate'].apply(lambda x: (t0 - x.max()).days)
frequency = data_cleaned.groupby('CustomerID')['InvoiceNo'].nunique()
monetary_value = data_cleaned.groupby('CustomerID')['TotalPrice'].sum()

# Теперь сгруппируем данные по идентификатору клиента и рассчитаем агрегированные характеристики
rfm_table = pd.DataFrame({'Recency': recency, 'Frequency': frequency, 'MonetaryValue': monetary_value})

rfm_table.shape

```

После того как мы подготовили RFM-таблицу, давайте визуализируем каждую из трёх компонент, например, в виде коробчатых диаграмм (boxplot).

```python
boxes = [px.box(rfm_table, x=column) for column in rfm_table.columns]
fig = make_subplots(rows=1, cols=3, subplot_titles=("Recency","Frequency", "MonetaryValue"))

for i, box in enumerate(boxes):
    fig.add_trace(boxes[i]['data'][0], row=1, col=i+1)

fig.update_layout(showlegend=True)
#fig.write_html('plotly/boxplot_1.html')
fig.show()
```

Что интересного здесь можно увидеть? Есть клиенты с аномально большим количеством сделанных заказов (более 100 штук), а также клиенты, общая стоимость заказов которых превышает 190 тысяч фунтов стерлингов.

Чем это плохо? Выбросы могут отрицательно сказаться на результатах работы методов кластеризации, неустойчивых к ним, например алгоритма KMeans, поэтому хотелось бы от них избавиться. Однако терять много ценных данных о клиентах тоже не хочется, поэтому ограничимся верхней границей соответствующей квантили уровня 0.95. Таким образом, мы удалим данные тех клиентов, для которых значение параметра Frequency или параметра Monetary Value выше, чем у 95 % клиентов.


Удалим из rfm-table записи о клиентах для которых выполняется хотя бы одно из условий - Frequency или Monetary Value выше, чем у 95 % клиентов.

Frequency_0.95 и MonetaryValue_0.95 - это будут квантили уровня 0.95 для этих признаков

```python
# Применим квантили для признаков Frequency и MonetaryValue
frequency_095 = rfm_table['Frequency'].quantile(0.95)
monetary_value_095 = rfm_table['MonetaryValue'].quantile(0.95)

# Используем метод маскирования loc[] для удаления таких записей из таблицы
rfm_table_cleaned = rfm_table.loc[(rfm_table['Frequency'] <= frequency_095) & (rfm_table['MonetaryValue'] <= monetary_value_095)]

rfm_table_cleaned.shape
```

Теперь давайте взглянем снова на наши коробчатые диаграммы.

```python
boxes = [px.box(rfm_table_cleaned, x=column) for column in rfm_table_cleaned.columns]
fig = make_subplots(rows=1, cols=3, subplot_titles=("Recency","Frequency", "MonetaryValue"))

for i, box in enumerate(boxes):
    fig.add_trace(boxes[i]['data'][0], row=1, col=i+1)

fig.update_layout(showlegend=True)
#fig.write_html('plotly/boxplot_1.html')
fig.show()
```

## 3. Моделирование и оценка качества моделей

### 3.1.1 Кластеризация на основе RFM-характеристик. Часть 1

Сформулируем наши дальнейшие шаги по кластеризации.

Чтобы результаты кластеризации было удобнее интерпретировать, мы можем уменьшить размерность исходных признаков до двух компонент.

**Подсказка.** Чтобы методы понижения размерности работали стабильно, данные необходимо стандартизировать/нормализовать. Для удобства можно обернуть эти шаги по предобработке данных в pipeline.

Произведем предобработку исходных данных. На основе RFM-признаков кластеризуем клиентов онлайн-магазина подарков с помощью известных нам методов (используем 5 методов).

Подберем оптимальное количество кластеров для выбранных методов с помощью коэффициента силуэта, перебирая возможные значения от 3 до 10 включительно (большее или меньшее количество кластеров будет нерелеватно для маркетинга). 

Выберем метод с наибольшим коэффициентом силуэта, сделаем предсказание меток кластеров.

Итак, для начала воспользуемся методом главных компонент (PCA) для уменьшения размерности до двух компонент. Этот метод способен найти новые независимые переменные (главные компоненты), которые объяснят наибольшую часть дисперсии исходных данных. PCA позволит сократить размерность данных, сохраняя при этом наибольшую часть информации. Ну а для стабильности работы метода применим стандартизацию над данными и обернем все в pipeline.

```python
# Создадим pipeline c шагами по обучению и предобработке данных и PCA
pipeline_pca = Pipeline([('scaler', StandardScaler()), ('pca', PCA(n_components=2))])

# Теперь применим pipeline к нашим rfm-данным 
pca_transformed = pipeline_pca.fit_transform(rfm_table_cleaned)

# Создадим DataFrame с преобразованными данными после применения PCA
pca_transformed = pd.DataFrame(pca_transformed, columns=['pc1', 'pc2'])


display(pca_transformed)
```

После применения метода главных компонент, мы можем использовать полученные две компоненты для визуализации данных используя scatter plot. В этом типе графика каждая точка представит собой наблюдение в пространстве двух компонент.

```python
# Создадим наш scatter plot
sns.scatterplot(data=pca_transformed, x='pc1', y='pc2', palette='Set1')

# Настроим оси и их заголовки
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title('Scatter Plot после PCA')

# Посмотрим на наш scatter plot
plt.show()
```

Найдем долю дисперсии исходных данных которая объясняет первая главная компонента, а какую долю вторая.

Для определения долей дисперсии можно воспользоваться атрибутом explained_variance_ratio_ объекта PCA

```python
# Получим доли дисперсии
explained_variance_ratio = pipeline_pca['pca'].explained_variance_ratio_

print("Доля дисперсии, объясняемая первой главной компонентой:", round(explained_variance_ratio[0], 2))
print("Доля дисперсии, объясняемая второй главной компонентой:", round(explained_variance_ratio[1], 2))
```

Попробуем разделить наше пространство главных компонет на сегменты. Для этого нужно будет применить методы кластеризации.

**Первый метод давайте возьмем K-means**. Этот метод разбивает данные на заранее заданное количество кластеров, определяя центроиды и минимизируя сумму квадратов расстояний от каждой точки до ближайшего центроида.

Нужно будет подобрать оптимальное количество кластеров для метода k-means с помощью коэффициента силуэта, перебирая возможные значения от 3 до 10 включительно.

```python
# Подберем оптимальное кол-во кластеров
num_clusters_range = range(3, 11)
silhouette_scores = []

for num_clusters in num_clusters_range:
    # Создаем экземпляр модели KMeans
    kmeans = KMeans(n_clusters=num_clusters, random_state=42)
    cluster_labels = kmeans.fit_predict(pca_transformed)
    score = silhouette_score(pca_transformed, cluster_labels)
    silhouette_scores.append(score)
    
for num_clusters, silhouette_score in zip(num_clusters_range, silhouette_scores):
    print(f"Кол-во кластеров: {num_clusters}, Оценка силуэта: {round(silhouette_score, 2)}")
```

Выбор оптимального количества кластеров можно основывать на наивысшем значении коэффициента силуэта, что указывает на лучшую структуру кластеров. Следовательно метод кластеризации k-means и коэффициент силуэта нам указывают на 3 кластера.

**Второй метод давайте возьмем метод иерархической кластеризации (Agglomerative Clustering)**. Соответственно исходя из названия этот метод использует иерархическую стратегию для кластеризации. Он начинает с каждой точки в отдельном кластере и последовательно объединяет ближайшие кластеры, пока не получится один общий кластер.

Нужно будет также подобрать оптимальное количество кластеров для метода Agglomerative Clustering с помощью коэффициента силуэта, перебирая возможные значения от 3 до 10 включительно.

```python
# Подберем оптимальное кол-во кластеров
num_clusters_range = range(3, 11)
silhouette_scores = []

for num_clusters in num_clusters_range:
    # Создаем экземпляр модели Agglomerative Clustering
    agg = AgglomerativeClustering(n_clusters=num_clusters)
    agg_labels = agg.fit_predict(pca_transformed)
    
    
    score = silhouette_score(pca_transformed, agg_labels)
    silhouette_scores.append(score)
    
    print(f"Кол-во кластеров: {num_clusters}, Оценка силуэта: {round(score, 2)}")
```

Метод иерархической кластеризации (Agglomerative Clustering)  и коэффициент силуэта нам указывают на 3 кластера.

**Третий метод давайте возьмем метод GMM (Gaussian Mixture Models)**. Это вероятностная модель, представляющая каждый кластер как гауссовское распределение. Он моделирует данные как смесь гауссовских распределений и использует EM-алгоритм для оценки параметров и кластеризации данных.

Также нужно будет подобрать оптимальное количество кластеров для метода GMM с помощью коэффициента силуэта, перебирая возможные значения от 3 до 10 включительно.

```python
# Подберем оптимальное кол-во кластеров
num_clusters_range = range(3, 11)
silhouette_scores = []

for num_clusters in num_clusters_range:
    # Создаем экземпляр модели GMM
    gmm = GaussianMixture(n_components=num_clusters, random_state=42)
    gmm_labels = gmm.fit_predict(pca_transformed)
    
    
    score = silhouette_score(pca_transformed, gmm_labels)
    silhouette_scores.append(score)
    
    print(f"Кол-во кластеров: {num_clusters}, Оценка силуэта: {round(score, 2)}")
```

Метод GMM и коэффициент силуэта нам указывают на 3 кластера.

**Четвертый метод давайте возьмем метод Spectral Clustering**. Этот метод основан на спектральной теории графов. Он преобразует данные в графовое представление и выполняет кластеризацию, используя информацию о связях между точками.

Также нужно будет подобрать оптимальное количество кластеров для метода Spectral Clustering с помощью коэффициента силуэта, перебирая возможные значения от 3 до 10 включительно.

Есть один нюанс у метода Spectral Clustering. Он вычислительно требовательный, особенно для больших наборов данных. Хоть мы и применили метод понижения размерности для наших rfm-данных, но этого не особо помогли и мы так и не дождались отработки метода Spectral Clustering. 

Можно использовать параллельные вычисления, но нам они не подойдут, так как все же железо, на котором запускается метод не подойдет для этого и это будет также длительное ожидание.

Попробуем использовать параметр affinity, который определяет способ вычисления матрицы близости. Если изменить значение этого параметра на nearest_neighbors, то мы сможем вычислить матрицу ближайших соседей, вместо полной матрицы близости. Что в свою очередь значительно должно сократить процессорное время.

```python
# Подберем оптимальное кол-во кластеров
num_clusters_range = range(3, 11)
silhouette_scores = []

for num_clusters in num_clusters_range:
    # Создаем экземпляр модели Spectral Clustering
    spectral_clustering = SpectralClustering(n_clusters=num_clusters, random_state=42, affinity='nearest_neighbors')
    cluster_labels = spectral_clustering.fit_predict(pca_transformed)
    
    score = silhouette_score(pca_transformed, cluster_labels)
    silhouette_scores.append(score)
    
    print(f"Кол-во кластеров: {num_clusters}, Оценка силуэта: {round(score, 2)}")
```

Отлично, метод отработал примерно за 3 секунду. Метод Spectral Clustering и коэффициент силуэта нам указывают на 3 кластера.

**Пятый метод давайте возьмем метод Birch**. Этот метод иерархической кластеризации строит иерархическую структуру, используя сжатое представление данных. Он позволяет быстро и эффективно кластеризовать большие объемы данных.

Также нужно будет подобрать оптимальное количество кластеров для метода Birch с помощью коэффициента силуэта, перебирая возможные значения от 3 до 10 включительно. Метод Birch имеет параметр threshold, с который можно экспериментировать для контроля компактности кластеров.

```python
# Подберем оптимальное кол-во кластеров
num_clusters_range = range(3, 11)
silhouette_scores = []

for num_clusters in num_clusters_range:
    # Создаем экземпляр модели Birch
    birch = Birch(n_clusters=num_clusters, threshold=0.5)
    cluster_labels = birch.fit_predict(pca_transformed)
    
    score = silhouette_score(pca_transformed, cluster_labels)
    silhouette_scores.append(score)
    
for num_clusters, silhouette_score in zip(num_clusters_range, silhouette_scores):    
    print(f"Кол-во кластеров: {num_clusters}, Оценка силуэта: {round(silhouette_score, 2)}")
```

В целом и без параметра threshold у нас получались такие же значения оценки силуэта. Метод Birch и коэффициент силуэта нам указали на 8 кластеров.

Мы применили 5 моделей кластеризации для подбора кол-ва кластеров с помощью коэффициента силуэта и теперь можно сделать вывод.
Наибольшую оценку силуэта показал метод K-means. Следовательно дальнейший расчет и  визуализацию наших данных мы будем проводить для трех кластеров и с помощью этого метода.

#### 3.1.2.2 Построение профиля кластеров (полярная диаграмма)

Чтобы результаты было проще интерпретировать, давайте познакомимся с одним из способов визуализации профиля кластеров — **Radar Chart** (полярная диаграмма, или диаграмма паутины). Это графическое представление значений нескольких эквивалентных категорий в форме паутины.

Radar Chart часто используется при определении профиля кластеров. На концах паутины откладываются оси, соответствующие признакам, описывающим объекты. На каждой из осей для каждого кластера откладываются средние значения соответствующих характеристик. Соединив точки по осям, мы получаем многоугольник. 

Пример полярной диаграммы для задачи кластеризации учеников по интересам:

На этой диаграмме мы видим визуализацию признаков для одного из кластеров. Видно, что ученики, принадлежащие этому кластеру, в большей степени увлекаются музыкой (Music), а в меньшей — программированием (Programm).

В модуле `graph_objects` библиотеки `plotly` есть встроенная функция `Scatterpolar`, которая позволяет построить полярную диаграмму. На основе этой функции мы подготовили для вас функцию `plot_cluster_profile()`, которая позволяет визуализировать профиль каждого из кластеров в виде полярной диаграммы. У неё есть два параметра: `grouped_data` — сгруппированные по кластерам характеристики объектов (клиентов), `n_clusters` — количество кластеров.

Главное условие использования полярной диаграммы — все признаки должны быть приведены к единому масштабу с помощью нормализации, где 1 будет означать максимум, а 0 — минимум. Шаг с нормализацией мы также добавили в функцию `plot_cluster_profile()`.


```python
def plot_cluster_profile(grouped_data, n_clusters):
    """Функция для визуализации профиля кластеров в виде полярной диаграммы.

    Args:
        grouped_data (DataFrame): таблица, сгруппированная по номерам кластеров с агрегированными характеристиками объектов.
        n_clusters (int): количество кластеров
    """
    # Нормализуем сгруппированные данные, приводя их к масштабу 0-1.
    scaler = preprocessing.MinMaxScaler()
    grouped_data = pd.DataFrame(scaler.fit_transform(grouped_data), columns=grouped_data.columns)
    # Создаём список признаков
    features = grouped_data.columns
    # Создаём пустую фигуру
    fig = go.Figure()
    # В цикле визуализируем полярную диаграмму для каждого кластера
    for i in range(n_clusters):
        # Создаём полярную диаграмму и добавляем её на общий график
        fig.add_trace(go.Scatterpolar(
            r=grouped_data.iloc[i].values, # радиусы
            theta=features, # название засечек
            fill='toself', # заливка многоугольника цветом
            name=f'Cluster {i}', # название — номер кластера
        ))
    # Обновляем параметры фигуры
    fig.update_layout(
        showlegend=True, # отображение легенды
        autosize=False, # устаналиваем свои размеры графика
        width=800, # ширина (в пикселях)
        height=800, # высота (в пикселях)
    )
    # Отображаем фигуру
    fig.show()
```

Воспользуемся готовым инструментом для анализа профиля кластеров, чтобы проинтерпретировать результаты, полученные на предыдущем шаге.

```python
# Укажем функции наши сгруппированные данные 
plot_cluster_profile(cluster_grouped, n_clusters=3)
```

Мы производили кластеризацию в пространстве главных компонент (метод PCA) и с помощью RFM-анализа нам удалось выделить три сегмента клиентов. **Второй метод также будет метод иерархической кластеризации (Agglomerative Clustering)**. 
Нужно будет также подобрать оптимальное количество кластеров для метода Agglomerative Clustering с помощью коэффициента силуэта, перебирая возможные значения от 3 до 10 включительно.

```python
# Подберем оптимальное кол-во кластеров
num_clusters_range = range(3, 11)
silhouette_scores = []

for num_clusters in num_clusters_range:
    # Создаем экземпляр модели Agglomerative Clustering
    agg = AgglomerativeClustering(n_clusters=num_clusters)
    agg_labels = agg.fit_predict(tsne_transformed)
    
    
    score = silhouette_score(tsne_transformed, agg_labels)
    silhouette_scores.append(score)
    
    print(f"Кол-во кластеров: {num_clusters}, Оценка силуэта: {round(score, 2)}")
```

Метод иерархической кластеризации (Agglomerative Clustering) и коэффициент силуэта нам указывают также на 7 кластеров.

**Третим методом также будет метод GMM (Gaussian Mixture Models)**.
Также нужно будет подобрать оптимальное количество кластеров для метода GMM с помощью коэффициента силуэта, перебирая возможные значения от 3 до 10 включительно.

```python
# Подберем оптимальное кол-во кластеров
num_clusters_range = range(3, 11)
silhouette_scores = []

for num_clusters in num_clusters_range:
    # Создаем экземпляр модели GMM
    gmm = GaussianMixture(n_components=num_clusters, random_state=42)
    gmm_labels = gmm.fit_predict(tsne_transformed)
    
    
    score = silhouette_score(tsne_transformed, gmm_labels)
    silhouette_scores.append(score)
    
    print(f"Кол-во кластеров: {num_clusters}, Оценка силуэта: {round(score, 2)}")
```

Метод GMM и коэффициент силуэта нам указывают на 5 кластеров.

**Четвертый методом будет метод Spectral Clustering с параметром affinity = nearest_neighbors**.
Также нужно будет подобрать оптимальное количество кластеров для метода Spectral Clustering с помощью коэффициента силуэта, перебирая возможные значения от 3 до 10 включительно.

```python
# Подберем оптимальное кол-во кластеров
num_clusters_range = range(3, 11)
silhouette_scores = []

for num_clusters in num_clusters_range:
    # Создаем экземпляр модели Spectral Clustering
    spectral_clustering = SpectralClustering(n_clusters=num_clusters, random_state=42, affinity='nearest_neighbors')
    cluster_labels = spectral_clustering.fit_predict(tsne_transformed)
    
    score = silhouette_score(tsne_transformed, cluster_labels)
    silhouette_scores.append(score)
    
    print(f"Кол-во кластеров: {num_clusters}, Оценка силуэта: {round(score, 2)}")
```

Метод Spectral Clustering и коэффициент силуэта нам указывают на 10 кластеров. Причем мы видим, что тут появилось для некоторых кластеров отрицательное значение оценки силуэта.

**Ну и пятый метод у нас метод Birch**. 
Также нужно будет подобрать оптимальное количество кластеров для метода Birch с помощью коэффициента силуэта, перебирая возможные значения от 3 до 10 включительно.

```python
# Подберем оптимальное кол-во кластеров
num_clusters_range = range(3, 11)
silhouette_scores = []

for num_clusters in num_clusters_range:
    # Создаем экземпляр модели Birch
    birch = Birch(n_clusters=num_clusters, threshold=0.5)
    cluster_labels = birch.fit_predict(tsne_transformed)
    
    score = silhouette_score(tsne_transformed, cluster_labels)
    silhouette_scores.append(score)
    
for num_clusters, silhouette_score in zip(num_clusters_range, silhouette_scores):    
    print(f"Кол-во кластеров: {num_clusters}, Оценка силуэта: {round(silhouette_score, 2)}")
```

Метод Birch и коэффициент силуэта нам указывают также на 7 кластеров.

Мы применили 5 моделей кластеризации для подбора кол-ва кластеров с помощью коэффициента силуэта и теперь можно сделать вывод. Удивительно, но на этот раз 3 метода показали одинаковый результат оценки силуэта = 0.6100000143051147. Поэтому тут можно выбрать любой из трех методов. Благодаря алгоритму t-SNE мы достигли нужного кол-ва кластеров и поэтому дальнейшие расчеты и визуализацию наших данных мы будем проводить для семи кластеров.
